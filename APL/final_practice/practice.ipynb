Pandas and Numpy

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

Making Data in respective frame

```python
pd_data = pd.read_csv('./covid.csv')
print(pd_data)
```

accusing the data of specific row using idex ==> iloc and loc using boolean 

```python
pd_data.loc[pd_data['date_died']!='9999-99-99']
pd_data.iloc[0]
```

Counting the data

```python
pd_data['sex'].unique()
pd_data['sex'].value_counts()
pd_data['sex'].replace(['Female','Male'],[2,1],inplace=True)
```

separating the value to died and non dead people

```python
dead_people = pd_data.loc[pd_data['date_died']!='9999-99-99']
alive_patient = pd_data.loc[pd_data['date_died']=='9999-99-99']
print(dead_people)
```

graph of dead and alive patient

```python
f, ax = plt.subplots(figsize= (8,6))
ax= sb.countplot(x= 'sex',data=dead_people)
ax.set_xticklabels(['Male', 'Female'])
plt.show()
```
