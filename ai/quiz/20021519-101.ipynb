#### add a column named: [release_year] and plot a graph showing number of movies released in a year

```python
df['release_year'] = pd.to_datetime(df['Release Date']).dt.year

# Plot a bar graph to show the number of movies released in each year
df['release_year'].value_counts().plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Number of movies')
plt.title('Number of movies released in each year')
plt.show()

```
